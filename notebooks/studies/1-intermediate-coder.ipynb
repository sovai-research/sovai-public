````python
## Because Polars have no index, addition is harder
import polars as pl

# Create Polars DataFrames instead of Series
df3 = pl.DataFrame({
    "index": ['a', 'b', 'c', 'd'],
    "value": [1, 2, 3, 4]
})

df4 = pl.DataFrame({
    "index": ['d', 'c', 'b', 'a'],
    "value": [4, 3, 2, 1]
})

# Perform the addition
result = (
    df3.join(df4, on="index", how="outer", suffix="_df4")
    .with_columns([
        pl.col("value").fill_null(0),
        pl.col("value_df4").fill_null(0),
        (pl.col("value") + pl.col("value_df4")).alias("sum")
    ])
    .select(["index", "sum"])
    .sort("index")
)

print(result)
````

````python
import polars as pl

def get_loc(df, column, value):
    return df.with_row_index('row_number').filter(pl.col(column) == value).select('row_number').item()

# Assuming sp500_pl is your Polars DataFrame and 'Symbol' is the column containing the symbols

# Find the positions of 'MMM' and 'A'
i1 = get_loc(sp500_pl, 'Symbol', 'MMM')
i2 = get_loc(sp500_pl, 'Symbol', 'A')

# Print the locations of 'MMM' and 'A'
i1, i2
````

````python
# demonstrate concat with a specification of the
# stock tickets being part of the index
# this help disambiguate the duplicate dates using
# a hierarchal index (muli-index) which is important in finance
closes = pd.concat([msftA01[:3], aaplA01[:3]],
                    keys=['MSFT', 'AAPL'])
closes
````

````python
# demonstrate concatenation using two DataFrame's
# that each have two columns.  pandas will align the
# data in columns by the column names (labels)
msftAV = msft[['Adj Close', 'Volume']]
aaplAV = aapl[['Adj Close', 'Volume']]
pd.concat([msftAV, aaplAV])
````

````python
# demonstrate concatenation with DataFrame objects
# that do not have the same set of columns leading to NaN values

aaplA = aapl[['Adj Close']]
pd.concat([msftAV, aaplA])
````

````python
# perform an inner join on the DataFrame's
# since aaplA does not have a Volume column, pandas
# will not include that column in the result

pd.concat([msftAV, aaplA], join='inner')
````

````python
# concat along the rows, causing duplicate columns to
# be created in the result

msftA = msft[['Adj Close']]
closes = pd.concat([msftA, aaplA], axis=1)
closes[:3]
````

````python
# concat along rows using two DataFrame objects with
# different number of rows. This demonstrates how
# NaN values will be filled in those rows for AAPL
# which only has three rows compared to 5 for MSFT

pd.concat([msftAV[:5], aaplAV[:3]], axis=1,
          keys=['MSFT', 'AAPL'])
````

````python
# inner join can also be used along this axis
# this will not include rows with index labels that do...
# not exist in both DataFrame objects

pd.concat([msftA[:5], aaplA[:3]], axis=1,
          join='inner', keys=['MSFT', 'AAPL'])
````

````python
# ignore indexes and just concatenate the data and
# have the resultbe a default integer index

pd.concat([msftA[:3], aaplA[:3]], ignore_index=True)
````

##Up and down resampling of time-series

````python
# calculate the cumulative daily returns for MSFT
msft_cum_ret = (1 + (msftAC / msftAC.shift() - 1)).cumprod()
msft_cum_ret
````

````python
# resample to a monthly cumulative return
msft_monthly_cum_ret = msft_cum_ret.resample("M")
msft_monthly_cum_ret
````

````python
# verify the monthly average for 2012-01
msft_cum_ret['2017-01'].mean()
````

````python
# verify that the default resample techique is mean
msft_cum_ret.resample("M").mean()
````

````python
# resample to monthly and give us open, high, low, close
msft_cum_ret.resample("M").ohlc()[:5]
````

````python
# this will return an index with periods instead of timestamps
by_periods = msft_cum_ret.resample("M",kind="period").mean()
for i in by_periods.index[:5]:
    print("{0}:{1} {2}".format(i.start_time,
                                i.end_time,
                                by_periods[i]))
````

````python
# upsampling will be demonstrated using the second
# and third values (first is NaN)
sample = msft_cum_ret[1:3]
sample
````

````python
# upsampling this will have a lot of NaN's
by_hour = sample.resample("H")
by_hour
````
